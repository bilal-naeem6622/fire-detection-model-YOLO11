```python
model = YOLO('yolo11m.pt')
```

```python
train_results = model.train(
    data="/content/drive/My Drive/fire_detection/custom_data.yaml",
    epochs=100,                  # start lower, can increase based on validation
    imgsz=256,                   # match your actual image resolution
    patience=10,                 # if supported, helps stop early if no improvement
    batch=16                     # adjust based on your GPU RAM
)
```
